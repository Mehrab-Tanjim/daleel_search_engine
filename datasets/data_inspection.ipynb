```python
from langchain.docstore.document import Document
import os
import json
from tqdm import tqdm
import pickle 
import pandas as pd
```

```python
quran_dataset = pd.read_excel(r'archive\holy_quran_dataset.xlsx')
```

```python
quran_dataset.head()
```

```python
quran_dataset.columns
```

```python
quran_dataset[quran_dataset['surah_no'] == 102].iloc[-1]['ayah_en']
```

```python
merged_documents = []
page_content_column = 'EnglishTranslation'
meta_data_columns = ['SurahNameEnglish', 'SurahNo', 'AyahNo', 'ArabicText']

print("Loading data and converting to documents")
```

```python
for index, data_point in tqdm(quran_dataset.iterrows()):
    if not isinstance(data_point[page_content_column], float):
        metadata={}
        for meta_data_column in meta_data_columns:
            metadata[meta_data_column] = str(data_point[meta_data_column]).strip()

        merged_documents.append(Document(page_content=data_point[page_content_column].strip(), metadata=metadata))

print("Total number of ayats: ", len(merged_documents))
```

```python
# Save the merged_data as a pickle file
with open("quran_document_data.pkl", "wb") as pickle_file:
    pickle.dump(merged_documents, pickle_file)
```
